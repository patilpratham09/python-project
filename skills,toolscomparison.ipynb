```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_csv("C:/Users/Pratham Patil/Analysis data.csv")
```

```python
data = data[['title','skills','responsibility']]
data = data.dropna()
data.head()
```

```python
tools_languages = ["excel", "power bi", "sap", "sql", "python", "oracle", "ms office", "tableau", "microsoft office", "powerpoint"]
soft_skills = ["communication", "problem solving", "teamwork", "creativity", "adaptability", "critical thinking", "time management", "financial analysis", "reporting", "analysis", "modeling", "management", "planning"]
```

```python
def categorize_skills(skill_list):
    soft_skills_extracted = []
    tools_languages_extracted = []
    
    for skill in skill_list:
        skill_lower = skill.lower()
        if any(tool in skill_lower for tool in tools_languages):
            tools_languages_extracted.append(skill.strip())
        else:
            soft_skills_extracted.append(skill.strip())
    
    return soft_skills_extracted, tools_languages_extracted
```

```python
df = pd.DataFrame(data)
```

```python
df["skills_list"] = df["skills"].apply(lambda x: x.split("\n"))
df["soft_skills"], df["tools_languages"] = zip(*df["skills_list"].apply(categorize_skills))
```

```python
df["all_skills"] = df["soft_skills"].apply(lambda x: ' '.join(x)) + " " + df["tools_languages"].apply(lambda x: ' '.join(x))
```

```python
vectorizer = CountVectorizer().fit_transform(df["all_skills"])
vectors = vectorizer.toarray()
```

```python
cosine_sim = cosine_similarity(vectors)
```

```python
similarity_results = []

titles = df["title"].tolist()

for i, title1 in enumerate(titles):
    for j, title2 in enumerate(titles):
        if i != j:
            similarity_results.append(f"Similarity between {title1} and {title2}: {cosine_sim[i][j]:.2f}")
```

```python
for result in similarity_results:
    print(result)
```
