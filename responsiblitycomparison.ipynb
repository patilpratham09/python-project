```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/Pratham Patil/Analysis data.csv")
```

```python
Data = data[['title','industry/department','company','skills','responsibility']]
```

```python
Data.head()
```

```python
def clean_split_responsibilities(resp_str):
    if isinstance(resp_str, str):
        return [resp.strip().lower() for resp in resp_str.split('\n') if resp.strip()]
    return []
```

```python
data['responsibilities_clean'] = data['responsibility'].apply(clean_split_responsibilities)
```

```python
def calculate_similarity(list1, list2):
    return len(set(list1).intersection(set(list2)))
```

```python
similarity_scores = {}

for i, row_i in data.iterrows():
    for j, row_j in data.iterrows():
        if i >= j:  # Skip redundant pairs
            continue
        common_responsibilities = calculate_similarity(row_i['responsibilities_clean'], row_j['responsibilities_clean'])
        similarity_scores[(row_i['title'], row_j['title'])] = common_responsibilities
```

```python
for titles, score in similarity_scores.items():
    print(f"Similarity between {titles[0]} and {titles[1]}: {score}")
```
