```python
data = data[['title','skills','responsibility']]
data = data.dropna()
```

```python
data['skills'] = data['skills'].str.split('\n').apply(lambda x: [skill.strip() for skill in x])
```

```python
soft_skills_list = ['communication', 'teamwork', 'problem solving', 'leadership', 'time management', 'adaptability', 'creativity']
tools_languages_list = ['excel', 'sql', 'python', 'r', 'tableau', 'power bi', 'sap', 'microsoft office', 'javascript', 'java']
```

```python
def clean_split_skills(skills_str):
    if isinstance(skills_str, str):
        return [skill.strip().lower() for skill in skills_str.split('\n') if skill.strip()]
    return []
```

```python
def categorize_skills(skills):
    soft_skills = []
    tools_languages = []
    for skill in skills:
        if any(soft_skill in skill for soft_skill in soft_skills_list):
            soft_skills.append(skill)
        elif any(tool in skill for tool in tools_languages_list):
            tools_languages.append(skill)
    return soft_skills, tools_languages
```
